**Yukarıdaki Soru İçin Detaylı Çözüm :**

```python
# 1. Metni birleştir
all_text = " ".join(df["description"].dropna())

# 2. Metni kelimelere ayır ve temizle
words = re.findall(r'\w+', all_text.lower())  # Küçük harfe çevir ve kelimeleri ayır

# 3. Kelime frekanslarını hesapla
word_counts = Counter(words)

# 4. En çok geçen kelimeyi bul
most_common_word = word_counts.most_common(10)

print("En sık geçen kelime:", most_common_word)
```

### **1-En Büyük Değer ve İndeks Sırasını Bulma:**

### **2- DF Sürunları Üzerinde String Metotları Kullanımı:**

    Yöntem-1 -> round(df[df["type"] == "Movie"]["duration"].str.replace(" min","").astype(int).mean(),2) 

    Yöntem-2 -> round(df[df["type"] == "Movie"]["duration"].str.split().str[0].astype(int).mean(),2) 



### **3-Counter Kullanımı:**

    - from collections import Counter
 
    - Counter([c for s in categories for c in s]) -> Counter içerisinde list comprhensionyapısı ile for döngüsü kullanımı
 
    - most_common() -> Counter ile ifadeler sayıldıktan sonra maksimum yer alan değerleri bulma işlemi için kullanılan yapı






### **4-str.contains() Kullanımı:**

    - str.contains, bir sütundaki yazılarda belli bir kelime veya ifadeyi arar ve o satırda o kelime geçiyorsa "doğru" (True), geçmiyorsa "yanlış" (False) der. Genelde verileri süzmek için kullanılır.
 





### **5-str.split() Kullanımı:**
    
    - str.split, bir sütundaki yazıları belli bir ayraca göre parçalara böler ve bunları bir liste haline getirir. Ayracı sen belirlersin, mesela boşluk, virgül veya başka bir işaret.

    - df['Ali Veli'].str.split() : Default olarak ifadeleri boşluk karakterinden böler ve bir liste içerisinde ayrı ayrı elemanlar haline getirir.

    - ['Ali', 'Veli']

### 7-**count() - len() - size() Kullanımı:**

    - count: Sadece boş olmayan (NaN olmayan) değerlerin sayısını döner.
  
    - len: Bir sütundaki toplam eleman sayısını döner (NaN dahil).

    - size: DataFrame veya Series'teki toplam hücre sayısını döner (NaN dahil).



### **8-explode():**

    - explode, pandas'ta bir sütunda liste ya da benzeri yapıların olduğu durumlarda, her liste elemanını ayrı bir satıra bölmek için kullanılır. Yani bir satırı, içindeki elemanlar kadar satıra genişletir.

### **11-isna() Kullanımı:**

    - isna, pandas'ta bir DataFrame veya Series içinde eksik (NaN veya None) değerleri bulmak için kullanılan bir fonksiyondur. Bu fonksiyon, eksik değerler için True, dolu değerler için False döndürür.

    - isna, eksik (NaN) değerleri bulur ve True/False döner.

    - df.cast.isna().sum() == df.cast.isnull().sum


### **12-join() Kullanımı:**

    - join ifadesini Counter ile birlikte kullandığınızda, genellikle bir dizideki elemanların sıklığını hesaplamak için kullanılır. join, bir metin listesini birleştirerek tek bir metin haline getirmek için kullanılır ve ardından Counter bu metindeki kelimelerin veya elemanların sıklığını hesaplar.


### **13-to_datatime :**

    - to_datetime metodu, Pandas kütüphanesinde tarih ve saat verilerini dönüştürmek için kullanılır. Bu metot, çeşitli tarih ve saat biçimlerini standart bir biçime dönüştürerek tarih ve saat işlemlerini daha kolay ve tutarlı hale getirir.

    - df['date_added'] = pd.to_datetime(df['date_added'])
    
    - df['year_added'] = df['date_added'].dt.year
